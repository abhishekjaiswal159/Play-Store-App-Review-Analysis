## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Ensure 'Price' is a string before replacing '$'
play_store_df['Price'] = play_store_df['Price'].astype(str)

# Remove '$' and convert to float
play_store_df['Price'] = (
    play_store_df['Price']
    .str.replace('$', '', regex=True)  # Remove "$"
    .replace({'Everyone': '0'})  # Handle non-numeric values
)

# Convert to float (after ensuring all are valid numbers)
play_store_df['Price'] = pd.to_numeric(play_store_df['Price'], errors='coerce')

# Confirm the changes
print(play_store_df[['Price']].head(10))
````

### 🛠️ Data Wrangling Explanation

#### **🔍 Issue in the Original Dataset**
- The `Price` column contained a mix of numeric values with a **dollar sign (`$`)** and some **non-numeric values** like `"Everyone"`, making it a **string** instead of a numeric type.

#### **✅ Steps Taken to Fix It**
1. Converted the column to a string using `.astype(str)`.
2. Removed the **"$"** symbol using `.str.replace('$', '', regex=True)`.
3. Replaced non-numeric values (like `"Everyone"`) with `"0"` to maintain consistency.
4. Converted the column to a numeric type using `pd.to_numeric(errors='coerce')`, which replaces invalid values with `NaN`.

#### **⚠️ Why Did We Get All `0.0` Values in the Output?**
- Most apps in the dataset are **free**, so `Price = 0` is the most common value.
- Any **non-numeric values** (like `"Everyone"`) were replaced with `NaN`, making them **invisible** in the first few rows of the output.
- Some paid app prices may exist, but they were not in the first 10 rows displayed.

#### **📌 Next Steps for Validation**
To ensure the conversion worked correctly, we will:
1. **Check unique values in `Price`**: `play_store_df['Price'].unique()`
2. **Identify missing values**: `play_store_df[play_store_df['Price'].isna()]`
3. **Verify non-zero prices exist**: `play_store_df[play_store_df['Price'] > 0].head(10)`

This explanation helps us confirm that the data transformation was successful and detect any remaining inconsistencies before further analysis.


````python
# Handling Missing Values in Play Store Dataset
play_store_df = play_store_df.copy()  # Ensure modifications happen on the original DataFrame

# Fill missing numerical values with median
play_store_df['Rating'] = play_store_df['Rating'].fillna(play_store_df['Rating'].median())

# Fill missing categorical values with mode
play_store_df['Type'] = play_store_df['Type'].fillna(play_store_df['Type'].mode()[0])
play_store_df['Content Rating'] = play_store_df['Content Rating'].fillna(play_store_df['Content Rating'].mode()[0])
play_store_df['Current Ver'] = play_store_df['Current Ver'].fillna(play_store_df['Current Ver'].mode()[0])
play_store_df['Android Ver'] = play_store_df['Android Ver'].fillna(play_store_df['Android Ver'].mode()[0])

# Verify if missing values are handled
print("🔹 Missing Values after Handling in Play Store Dataset:")
print(play_store_df.isnull().sum())

# Handling Missing Values in User Reviews Dataset
user_reviews_df = user_reviews_df.copy()  # Ensure modifications happen on the original DataFrame

# Drop rows where 'Translated_Review' is missing (as it affects sentiment analysis)
user_reviews_df = user_reviews_df.dropna(subset=['Translated_Review'])

# Fill missing sentiment values with mode
user_reviews_df['Sentiment'] = user_reviews_df['Sentiment'].fillna(user_reviews_df['Sentiment'].mode()[0])

# Fill missing numerical sentiment values with median
user_reviews_df['Sentiment_Polarity'] = user_reviews_df['Sentiment_Polarity'].fillna(user_reviews_df['Sentiment_Polarity'].median())
user_reviews_df['Sentiment_Subjectivity'] = user_reviews_df['Sentiment_Subjectivity'].fillna(user_reviews_df['Sentiment_Subjectivity'].median())

# Verify if missing values are handled
print("\n🔹 Missing Values after Handling in User Reviews Dataset:")
print(user_reviews_df.isnull().sum())
````

###Data Transformation & Feature Engineering

````python
# Ensure modifications happen on the original DataFrame
play_store_df = play_store_df.copy()

# Convert 'Installs' to numeric (remove '+' and ',')
play_store_df['Installs'] = (
    play_store_df['Installs']
    .str.replace(r"\+", "", regex=True)  # Remove '+'
    .str.replace(",", "", regex=True)    # Remove ','
    .replace("Free", "0")  # Convert 'Free' to 0
    .astype(float)  # Convert to float
)

# Convert 'Size' to a consistent unit (MB)
def convert_size(size):
    if 'M' in size:  # If already in MB, remove 'M' and convert to float
        return float(size.replace('M', ''))
    elif 'K' in size:  # Convert KB to MB
        return float(size.replace('K', '')) / 1024
    else:  # Handle cases like 'Varies with device'
        return None

play_store_df['Size'] = play_store_df['Size'].astype(str).apply(convert_size)

# Convert 'Last Updated' to DateTime format
play_store_df['Last Updated'] = pd.to_datetime(play_store_df['Last Updated'], errors='coerce')

# Verify the changes
print("🔹 Data Types after Transformation:")
print(play_store_df.dtypes)

print("\n🔹 First 5 Rows after Transformation:")
print(play_store_df[['Installs', 'Size', 'Last Updated']].head())
````

###Data Cleaning: Reviews, Price & Handling Missing Values

````python
print("🔍 Checking 'Reviews' column values BEFORE cleaning:")
print(play_store_df['Reviews'].head(20).tolist())  # Print as a list for clear viewing
````

````python
import pandas as pd

# 🔹 Ensure modifications happen on the original DataFrame
play_store_df = play_store_df.copy()

# === 1️⃣ Clean the 'Reviews' Column (Proper Debugged Fix) ===
def clean_reviews(value):
    value = str(value).strip().replace(',', '')  # Remove commas and spaces
    if value == '' or value.lower() == 'nan':  # Handle empty values
        return None
    try:
        if value.endswith('M'):  # Convert 'M' to numeric (millions)
            return float(value.replace('M', '')) * 1e6
        return float(value)  # Convert normal numbers
    except ValueError:  # Catch any conversion errors
        print(f"❌ Skipping invalid value in 'Reviews': {value}")  # Debug print
        return None  # Return None for invalid values

play_store_df['Reviews'] = play_store_df['Reviews'].apply(clean_reviews)

# === 2️⃣ Clean the 'Price' Column ===
def clean_price(value):
    value = str(value).strip().replace('$', '')  # Remove spaces and '$'
    return pd.to_numeric(value, errors='coerce')  # Convert to float safely

play_store_df['Price'] = play_store_df['Price'].apply(clean_price)

# === 3️⃣ Handle Any Remaining Missing Values ===
# Fill missing numerical values with median
for col in ['Reviews', 'Size', 'Price']:
    if play_store_df[col].isnull().sum() > 0:
        play_store_df[col] = play_store_df[col].fillna(play_store_df[col].median())

# Fill missing categorical values with mode
for col in ['Type', 'Content Rating', 'Current Ver', 'Android Ver']:
    if play_store_df[col].isnull().sum() > 0:
        play_store_df[col] = play_store_df[col].fillna(play_store_df[col].mode()[0])

# 🔹 Verify the changes
print("✅ Data Types after Final Cleaning:")
print(play_store_df.dtypes)

print("\n✅ First 5 Rows after Final Cleaning:")
print(play_store_df[['Reviews', 'Price', 'Size', 'Type', 'Content Rating']].head())
````

#### Chart - 6 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# 🔹 Select relevant numerical columns for correlation
numeric_cols = ['Rating', 'Reviews', 'Size', 'Installs', 'Price']

# 🔹 Compute the correlation matrix
corr_matrix = play_store_df[numeric_cols].corr()

# 🔹 Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=14)
plt.show()
````

We selected the **Correlation Heatmap** to visually examine the strength and direction of relationships between key numerical variables such as `Rating`, `Reviews`, `Size`, `Installs`, and `Price`.

This chart helps us quickly identify:
- Which variables are strongly or weakly correlated.
- Positive or negative associations that might impact app performance.

By analyzing these correlations, we can make informed assumptions about which features are most relevant for app success. It provides a statistical foundation for understanding the dynamics between app characteristics.


##### 2. What is/are the insight(s) found from the chart?

### 📌 Insights Found from the Correlation Heatmap

- There is a **moderate positive correlation between the number of Installs and Reviews**, indicating that apps with more installs tend to receive more reviews, which makes logical sense as more users result in more feedback.
- **Rating has a weak correlation with other variables**, suggesting that user satisfaction (rating) is influenced by factors not directly represented in numeric form (e.g., UI/UX, customer support, app stability).
- **Price is negatively correlated with Installs**, indicating that **free apps are more likely to be downloaded** compared to paid apps.

These insights highlight user preferences and can be instrumental in designing app strategies.
